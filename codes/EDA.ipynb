### This Notebook contains code for Exploratory Data Analysis done on the data set under study

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
clean_data = pd.read_csv("../Data/Data_File_csv.csv")
```

```python
nan_count = clean_data.isna().sum()
```

```python
mod_data_1 = clean_data.loc[:,nan_count<0.1*clean_data.shape[0]]
mod_data_1
```

```python
cor_matrix = mod_data_1.corr()
cor_matrix
```

```python
mod_data_1 = mod_data_1.replace("Yes", 1)
mod_data_1 = mod_data_1.replace("Maybe", 1)
mod_data_1 = mod_data_1.replace("No", 0)
```

```python
temp = country_dat[country_dat[('Current_Mental_Illness', 'count')] >10]
temp[[('Current_Mental_Illness', 'mean'),('Diagnosed_Mental_condition_Med_prof', 'mean')]].plot.bar(title="Country Effect")
plt.ylabel("Category Percentage")
ax = plt.gca()
plt.legend(bbox_to_anchor=(1, -0.91), bbox_transform=ax.transAxes)
```

```python
work_type = mod_data_1[["Work_Remotely", "Current_Mental_Illness", "Diagnosed_Mental_condition_Med_prof"]].groupby(["Work_Remotely"]).agg(["sum", "count", "mean"])
work_type[[('Current_Mental_Illness', 'mean'),('Diagnosed_Mental_condition_Med_prof', 'mean')]].plot.bar(title="Remote Work Effect")
plt.ylabel("Category Percentage")
ax = plt.gca()
plt.legend(bbox_to_anchor=(1.1, 1.1), bbox_transform=ax.transAxes)
```

```python
gender_dat = mod_data_1[["Gender", "Current_Mental_Illness", "Diagnosed_Mental_condition_Med_prof"]].groupby(["Gender"]).agg(["sum", "count", "mean"])
gender_dat[[('Current_Mental_Illness', 'mean'),('Diagnosed_Mental_condition_Med_prof', 'mean')]].plot.bar(title="Gender Effect")
plt.ylabel("Category Percentage")
ax = plt.gca()
plt.legend(bbox_to_anchor=(1.5, 1), bbox_transform=ax.transAxes)
```

```python
mod_data_1["Gender"] = mod_data_1["Gender"].str.upper()
mod_data_1["Gender"] = mod_data_1["Gender"].str.strip()
```

```python
mod_data_1["Current_Mental_Illness"] = mod_data_1["Current_Mental_Illness"].replace("Yes", 1)
mod_data_1["Current_Mental_Illness"] = mod_data_1["Current_Mental_Illness"].replace("Maybe", 1)
mod_data_1["Current_Mental_Illness"] = mod_data_1["Current_Mental_Illness"].replace("No", 0)

mod_data_1["Diagnosed_Mental_condition_Med_prof"] = mod_data_1["Diagnosed_Mental_condition_Med_prof"].replace("Yes", 1)
mod_data_1["Diagnosed_Mental_condition_Med_prof"] = mod_data_1["Diagnosed_Mental_condition_Med_prof"].replace("Maybe", 1)
mod_data_1["Diagnosed_Mental_condition_Med_prof"] = mod_data_1["Diagnosed_Mental_condition_Med_prof"].replace("No", 0)
```
